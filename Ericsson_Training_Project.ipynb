```python
import numpy as np
import pandas as pd
```

```python
RSRP = ('/content/RSRP.csv')
dataset = pd.read_csv(RSRP)
```

```python
# total count of NaN values
print(dataset.isnull().sum().sum())
```

```python
d_drop=dataset.dropna(axis = 0, how ='any')
```

```python
d_drop
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
corr_matrix = d_drop.corr().abs()
```

```python
threshold = 0.01
```

```python
filtered_corr_df = corr_matrix[(corr_matrix >= threshold) & (corr_matrix != 1.000)]

plt.figure(figsize=(7,7))
sn.heatmap(filtered_corr_df, annot=True, cmap="Reds")
plt.show()
```

```python
import pyproj
import shapely
import shapely.ops as ops
from shapely.geometry.polygon import Polygon
from functools import partial
```

```python
geom = Polygon([(24.54054, 46.784737), (24.612879, 46.857464), (24.813457, 46.742294)])
geom_area = ops.transform(
    partial(
        pyproj.transform,
        pyproj.Proj(init='EPSG:4326'),
        pyproj.Proj(
            proj='aea',
            lat_1=geom.bounds[1],
            lat_2=geom.bounds[3]
        )
    ),
    geom)
#ALL C OPERATOR
# Print the area in km^2
print(geom_area.area/1000000)
```
